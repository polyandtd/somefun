```python
import pandas as pd
```

```python
df = pd.read_csv('milk_promo_sales.csv')
df.head()
```

```python
df.sort_values(['store_id', 'period_id'], inplace=True)
df.head()
```

```python
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
def rating(row):
    global current_rating
    global previous_period_id
    global previous_store_id
    
    current_rating += 1
    
    if (row.period_id - previous_period_id > 1) or (row.store_id != previous_store_id):
        current_rating = 0
    
    previous_period_id = row.period_id
    previous_store_id = row.store_id

    return current_rating
```

```python
current_rating = -1
previous_store_id = df.loc[0, 'store_id']
previous_period_id = df.loc[0, 'period_id']

df['promo_cw'] = df.apply(rating, axis=1)
df.head()
```

```python
def rating_id(row):
    global current_rating_id
    global previous_store_id
    
    current_rating_id += 1
    
    if row.promo_cw != 0:
        current_rating_id -=1
    if row.store_id != previous_store_id:
        current_rating_id = 1
    
    previous_store_id = row.store_id

    return current_rating_id
```

```python
current_rating_id = 0
previous_store_id = df.loc[0, 'store_id']

df['promo_id'] = df.apply(rating_id, axis=1)
df.head(30)
```
